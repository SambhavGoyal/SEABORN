```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x= 'sex', data=tips)
```

```python
sns.set_style('ticks')
```

```python
sns.countplot(x= 'sex', data=tips)
sns.set_style('whitegrid')
```

```python
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
```

```python
sns.countplot(x= 'sex', data=tips)
sns.set_style('ticks')
```

```python
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
sns.despine()
```

```python
plt.figure(figsize = (12,5))
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
```

```python
# Scale and Context
sns.set_context('poster')
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
```

```python
sns.set_context('notebook')
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
```

```python
sns.set_context('paper')
sns.countplot(x= 'sex', data=tips)
sns.set_style('dark')
```
